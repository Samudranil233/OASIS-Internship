```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime as dt
import calendar
```

```python
data=pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
data.isnull().sum()
```

```python
data.rename(columns={'Region':'States',
                     ' Date':'Date',
                     ' Estimated Unemployment Rate (%)':'Estimated Unemployment Rate (%)',
                     ' Estimated Employed':'Estimated Employed',
                     ' Estimated Labour Participation Rate (%)':'Estimated Labour Participation Rate (%)',
                     'Region.1':'Region'},inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.drop(columns=[' Frequency'],inplace=True)
```

```python
df = data
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation=35)
sns.barplot(data=df, x='Region', y='Estimated Employed',palette='Set2',order=df.groupby('Region')['Estimated Employed'].mean().sort_values().index)
plt.tight_layout()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='States', y='Estimated Unemployment Rate (%)', order=df.groupby('States')['Estimated Unemployment Rate (%)'].mean().sort_values().index)
plt.xlabel('States')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate by States (Sorted)')
plt.xticks(rotation=65)  
plt.tight_layout()
plt.show()
```
